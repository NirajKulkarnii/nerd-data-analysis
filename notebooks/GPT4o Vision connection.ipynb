```python
def generate_chat_completion_from_folder(b64_image, prompt, model='gpt-4o-mini', max_tokens=300):
    """
    Sends a request to the OpenAI API to generate a chat completion based on the given prompt
    and images from a specified folder.

    Args:
        folder_path (str): The path to the folder containing images.
        prompt (str): The text prompt to be used by the model.
        model (str): The model name to be used. Defaults to 'gpt-4o-mini'.
        max_tokens (int): The maximum number of tokens to generate. Defaults to 300.

    Returns:
        requests.Response: The response object from the OpenAI API.
    """
    api_key  = os.getenv('OPENAI_API_KEY')
    # Check if the API key is retrieved successfully
    if not api_key:
        raise ValueError("API key not found. Please set the environment variable.")
    # Define the headers for the request
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {api_key}"  
    }

    # Construct the final payload
    payload = {
        "model": model,
        "messages": [
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": prompt
                    },
                    {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{b64_image}"
                        }
                    }
                ] 
            }
        ],
        "max_tokens": max_tokens
    }
    
    # Send the POST request to the OpenAI API and return the response
    url = "https://api.openai.com/v1/chat/completions"
    response = requests.post(url, headers=headers, json=payload)
    return response.json()
```
